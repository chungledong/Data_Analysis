```python
# read data in file *.xls
def read_data(file_path, sheet_name=None):
    """
    Read data from an Excel file and return a DataFrame.
    """
    try:
        df = pd.read_excel(file_path, sheet_name=sheet_name)
        return df
    except Exception as e:
        logging.error(f"Error reading file {file_path}: {e}")
        sys.exit(1)
```

```python
def chuan_hoa_du_lieu(df, columns_to_normalize):
    """
    Chuẩn hóa dữ liệu trong các cột được chỉ định của DataFrame.

    Args:
        df (pd.DataFrame): DataFrame cần chuẩn hóa.
        columns_to_normalize (list): Danh sách tên các cột cần chuẩn hóa.

    Returns:
        pd.DataFrame: DataFrame đã được chuẩn hóa.
    """
    df_normalized = df.copy()  # Tạo bản sao để không sửa đổi DataFrame gốc

    for col in columns_to_normalize:
        if col in df_normalized.columns:
            # Kiểm tra kiểu dữ liệu của cột, chỉ áp dụng cho kiểu object (chuỗi)
            if df_normalized[col].dtype == 'object':
                # 1. Loại bỏ khoảng trắng thừa ở đầu và cuối chuỗi
                df_normalized[col] = df_normalized[col].str.strip()

                # 2. Chuẩn hóa khoảng trắng giữa các từ thành một khoảng trắng đơn
                df_normalized[col] = df_normalized[col].str.replace(r'\s+', ' ', regex=True)

                # 3. Loại bỏ các ký tự đặc biệt hoặc không mong muốn (có thể tùy chỉnh thêm)
                # Ví dụ: df_normalized[col] = df_normalized[col].str.replace(r'[^\w\s]', '', regex=True)

                # 4. Chuyển đổi thành chữ thường (tùy chọn, có thể hữu ích cho so sánh)
                # df_normalized[col] = df_normalized[col].str.lower()

                # 5. Viết hoa chữ cái đầu tiên của mỗi từ (tùy chọn, cho tên hoặc chức danh)
                # if col in ['Họ và tên', 'Vị trí công việc', 'Đơn vị công tác', 'Nơi đào tạo', 'Chuyên ngành']:
                #     df_normalized[col] = df_normalized[col].str.title()

    return df_normalized
```

```python
columns=['Mã nhân viên', 'Họ và tên', 'Giới tính', 'Trình độ đào tạo',
       'Vị trí công việc', 'Đơn vị công tác', 'Trạng thái', 'Trình độ đào tạo',
       'Nơi đào tạo', 'Chuyên ngành','Trạng thái']

data = chuan_hoa_du_lieu(data, columns)
```

```python
def chuan_hoa_cac_cot_ngay(df, column_day, target_format='%d/%m/%Y', errors='coerce'):
    """
    Chuẩn hóa nhiều cột ngày tháng năm trong DataFrame về định dạng chỉ định.

    Args:
        df (pd.DataFrame): DataFrame chứa các cột ngày tháng năm.
        column_day (list): Danh sách tên các cột chứa dữ liệu ngày tháng năm cần chuẩn hóa.
        target_format (str, optional): Định dạng ngày tháng năm mục tiêu (strftime format).
                                       Mặc định là '%Y-%m-%d' (YYYY-MM-DD).
        errors (str, optional): Cách xử lý lỗi khi chuyển đổi.
                                 - 'raise': Nếu gặp lỗi, sẽ phát sinh ngoại lệ.
                                 - 'coerce': Nếu gặp lỗi, sẽ chuyển giá trị không hợp lệ thành NaT.
                                 - 'ignore': Nếu gặp lỗi, sẽ giữ nguyên giá trị ban đầu.
                                 Mặc định là 'coerce'.

    Returns:
        pd.DataFrame: DataFrame với các cột ngày tháng năm đã được chuẩn hóa.
    """
    df_normalized = df.copy()

    for col in column_day:
        if col in df_normalized.columns:
            df_normalized[col] = pd.to_datetime(df_normalized[col], errors=errors)
            df_normalized[col] = df_normalized[col].dt.strftime(target_format)

    return df_normalized
```

```python
def loc_du_lieu_theo_trang_thai(df):
  """
  Lọc dữ liệu từ DataFrame chỉ giữ lại các dòng có trạng thái là 'Thử việc' hoặc 'Đang làm việc'.

  Args:
    df: DataFrame Pandas chứa cột 'Trạng thái'.

  Returns:
    DataFrame Pandas mới chỉ chứa các dòng có trạng thái 'Thử việc' hoặc 'Đang làm việc'.
    Trả về DataFrame rỗng nếu không có dòng nào thỏa mãn điều kiện.
  """
  cac_trang_thai_can_loc = ['Thử việc', 'Đang làm việc']
  df_da_loc = df[df['Trạng thái'].isin(cac_trang_thai_can_loc)]
  return df_da_loc
```
